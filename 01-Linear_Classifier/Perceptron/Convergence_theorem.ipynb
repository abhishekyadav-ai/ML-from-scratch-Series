```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Define the positive and negative classes
positive_points=np.array([(-2,3),(0,1),(2,-1)])
negative_points=np.array([(-2,1),(0,-1),(2,-3)])
```

```python
#Combine all points and labels
X=np.vstack((positive_points, negative_points))
y=np.array([1]*len(positive_points)+[-1]*len(negative_points))
R=0

R = max(np.linalg.norm(xi) for xi in X)
R
```

```python
#Perceptron algorithm
def perceptron(X,y,steps=10):
    theta=np.zeros(X.shape[1])
    theta0=0
    
    for j in range(steps):
        for i, xi in enumerate(X):
            if y[i]*(np.dot(xi,theta)+theta0)<=0:
                theta +=y[i]*xi
                theta0+=y[i]
                break       
        #Plot for each iteration
        #plot_decision_boundary(theta, theta0, j+1)
    return theta, theta0
 
    
    
```

```python
theta, theta0=perceptron(X,y,10)
theta_norm = np.linalg.norm(theta)

gamma = min(
    y[i] * (np.dot(X[i], theta) + theta0) / theta_norm
    for i in range(len(X))
)
print(f"Margin of the dataset is {gamma}")
atmost_error=(R/gamma)**2
print(F"Atmost the perceptron algorithm will make {atmost_error} mistakes for finding the best separator")
```
